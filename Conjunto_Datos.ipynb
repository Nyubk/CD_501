```python
#Limpiamos de datos NaN en caso de que estos hayan pasado desapercibidos y eliminamos duplicados
df_dActivity_copy.dropna(subset = 'TotalSteps')
df_dActivity_copy.dropna(subset = 'Calories')
df_dActivity_copy.dropna(subset = 'ActivityDate')
df_dActivity_copy['TotalSteps'].drop_duplicates
df_dActivity_copy
```

```python
#Generamos el dataframe con solos las columnas indicadas
df_dAtrain = pd.DataFrame(df_dActivity_copy.loc[:,["ActivityDate","TotalSteps","Calories"]])
df_dAtrain.DataFrame = df_dActivity_copy.loc[:,["ActivityDate","TotalSteps","Calories"]]
#Añadimos una cuarta columna que nos entregue la cantidad de pasos necesarios para perder al menos una caloría
df_dAtrain["Calories/TotalSteps"] = df_dActivity_copy["TotalSteps"]/ df_dActivity_copy["Calories"]
df_dAtrain.head(3)
```

```python
#Validamos la integridad de los datos
df_dAtrain.info()
```

```python
#Definir espacio de gráfica
fig, ax = plt.subplots(figsize=(8,6),dpi=100)

#Definir variables auxiliares
x = df_dAtrain['TotalSteps']
y = df_dAtrain['Calories']

p = ax.scatter(x, y ,
               s=30,
               cmap='RdBu_r',
               #vmin=colors.min)
               alpha=0.3,
               edgecolors='none')

#Barra de colores
cb = fig.colorbar(p, ax=ax, label='Calorias Quemadas ', extend='max')
cb.outline.set_visible(False)

ax.spines['right'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.set_facecolor('lightgray')
fig.tight_layout()

ax.set_title('Calorias quemadas/ Pasos Totales diarios')
ax.set_xlabel('Pasos totales por día')
ax.set_ylabel('Calorias')
```
